```python
data=pd.get_dummies(data,drop_first=True)#instead of this use scikit learn cozz dummies might cause some problems
```

```python
x=data.drop("Performance Index",axis=1)#defining feature and target value
y=data["Performance Index"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)#splitting data into train and test set
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Performace Index")
plt.ylabel("Predicted Performace Index")
plt.title("Actual vs Predicted")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R2 Score:", r_squared)
```
